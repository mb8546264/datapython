تأثیر سن و وضعیت شغلی روی درآمد

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# خواندن داده‌ها
file_path = 'sample_data.csv'  # مسیر فایل داده
data = pd.read_csv(file_path)
data.head()
data.info()
```

```python
data['is_employed'] = data['is_employed'].fillna('Unknown').astype(str)
data['income'] = data['income'].fillna(data['income'].median())
data['age'] = data['age'].fillna(data['age'].median())

```

```python
print("Unique values in 'is_employed':", data['is_employed'].unique())
#true:فرد شاغل است.
#false:فرد بیکار است.
#Unknown :وضعیت شغلی مشخص نیست.
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data,
    x='age',
    y='income',
    hue='is_employed',  
    style='is_employed',
    palette={'True': 'red', 'False': 'blue', 'Unknown': 'green'},  # رنگ‌های سفارشی
    markers={'True': 'o', 'False': 's', 'Unknown': 'D'},  # اشکال سفارشی
    alpha=0.7
)
plt.title('Scatter Plot: Income vs Age by Employment Status', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.legend(title='Employment Status', loc='upper left')
plt.show()
```
